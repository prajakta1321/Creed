```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer


df = pd.read_csv('GRIEVANCE_2000.csv')


df.dropna(subset=['dist_name'], inplace=True)


text_data = df['dist_name'].tolist()


tfidf_vectorizer = TfidfVectorizer()


tfidf_features = tfidf_vectorizer.fit_transform(text_data)


feature_names = tfidf_vectorizer.get_feature_names_out()


print("Vocabulary (Unique Words):")
print(feature_names)


print("\nTF-IDF Features:")
for i, document in enumerate(text_data):
    print(f"Document {i+1}:")
    for word, tfidf_score in zip(feature_names, tfidf_features.toarray()[i]):
        print(f"{word}: {tfidf_score:.4f}")
    print()
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import pandas as pd

df = pd.read_csv('GRIEVANCE_2000.csv')


df.dropna(subset=['dist_name'], inplace=True)


tokenized_text = [word_tokenize(text) for text in df['dist_name']]


word2vec_model = Word2Vec(sentences=tokenized_text, vector_size=100, window=5, min_count=1, workers=4)


word_to_check = input("Enter a word to check its vector and similar words: ")


if word_to_check in word2vec_model.wv.key_to_index:

    print("Vector for '{}'".format(word_to_check))
    print(word2vec_model.wv[word_to_check])


    similar_words = word2vec_model.wv.most_similar(word_to_check, topn=5)
    print("Similar words to '{}':".format(word_to_check))
    print(similar_words)
else:
    print("'{}' is not present in the vocabulary.".format(word_to_check))
```
