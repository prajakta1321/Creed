```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
X, _ = make_moons(n_samples=100, noise=0.04)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(9,3))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='inferno')
plt.title('DBSCAN Clustering')
plt.show()
```

# **Density-Based Spatial Clustering of Applications with Noise (DBSCAN)**

DBSCAN stands out as a density-based clustering algorithm. It doesn’t rely on predefining the number of clusters, making it suitable for various real-world scenarios.

The core idea behind DBSCAN is to group data points that are close to each other in terms of density while considering noise points that don’t belong to any cluster.

Handles irregularly shaped clusters and identifies noise points.
