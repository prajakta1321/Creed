```python
import pandas as pd
df= pd.read_csv("beer1.csv")

df.head()
```

```python
skewness = df.skew()
print("Skewness:")
print(skewness)
```

above skewness shows that final de semana is almost 0 which shows athat it is slightly positive for normal distrbition

```python
kurtosis = df.kurtosis()
print("\nKurtosis:")
print(kurtosis)
```

```python
import seaborn as sns

g = sns.displot(data=df, x="Final de Semana",kind ='kde',color='darkgreen')
for ax in g.axes.ravel():
  ax.text(x=0.97, y=0.97, transform=ax.transAxes, s="Skewness: %f" % df['Final de Semana'].skew(),fontsize=10, verticalalignment='top', horizontalalignment='right')
  ax.text(x=0.97, y=0.91, transform=ax.transAxes, s="Kurtosis: %f" % df['Final de Semana'].kurt(),fontsize=10, verticalalignment='top', horizontalalignment='right')

```
