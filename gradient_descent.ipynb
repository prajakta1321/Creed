Gradient Descent is a fundamental optimization algorithm used in machine learning and various other fields to minimize a function, typically a cost or loss function. It’s an iterative algorithm that adjusts the model’s parameters to find the minimum of the function, which represents the best possible values for those parameters.

The key component of gradient descent is the gradient (often denoted as ∇) of the objective function. The gradient points in the direction of the steepest increase in the function, so moving in the opposite direction will lead you closer to the minimum.

There are different variations of gradient descent, including:

1.Batch Gradient Descent

2.Stochastic Gradient Descent (SGD)

3.Mini-Batch Gradient Descent

4.Adaptive Methods


Gradient descent is a versatile optimization algorithm and is widely used in training machine learning models, especially neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

We define an example quadratic objective function f(x) = x^2 and its gradient f'(x) = 2x.

```python
def objective_function(x):
    return x**2
```

```python
def gradient(x):
    return 2 * x
```

The gradient_descent function performs the gradient descent optimization with the given learning rate and number of iterations.

We store the history of x values during each iteration to track the trajectory.

The code then plots the objective function and the trajectory of x values.

```python
def gradient_descent(learning_rate, iterations):
    x = 10  # Initial guess
    history = []
    for _ in range(iterations):
        history.append(x)
        x = x - learning_rate * gradient(x)

    return history
```

```python
learning_rate = 0.2
iterations = 25
```

```python
trajectory = gradient_descent(learning_rate, iterations)
```

```python
x = np.linspace(-12, 12, 400)
y = objective_function(x)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Objective Function')
plt.plot(trajectory, [objective_function(x) for x in trajectory], 'ro-', label='Trajectory')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
